# NYC Taxi Data Demo

This notebook loads and displays the classic Databricks NYC taxi dataset.

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("NYC Taxi Demo").getOrCreate()
```

```python
# Load the NYC taxi dataset from Databricks sample datasets
df = spark.read.format("delta").load("/databricks-datasets/nyctaxi/tables/nyctaxi_yellow")
```

```python
# Print schema to understand the data structure
print("Schema:")
df.printSchema()
```

```python
# Print first 10 rows
print("\nFirst 10 rows:")
df.show(10)
```
